```python
import pandas as pd
import numpy as np
import matplotlib.patches as patches
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("mycsvfile.csv")
```

```python
df
```

```python
df_fd_id_one = df.loc[df['f_id'] == 1]
df_fd_id_two = df.loc[df['f_id'] == 2]
```

```python
df_fd_id_one
```

```python
df_fd_id_two
```

```python
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)
```

```python
fig2 = plt.figure(figsize=(10,10),dpi=100)
ax2 = fig2.add_subplot(111, aspect='equal')
for index, row in df_fd_id_two.iterrows():
    ax2.add_patch(
        patches.Rectangle(
            (row["VST"], 40),
            3,
            5,
            fill=False 
        )
    )
for index, row in df_fd_id_one.iterrows():
    ax2.add_patch(
        patches.Rectangle(
            (row["VST"], 10),
            4,
            5,
            fill=False 
        )
    )

plt.xlim(-10, 50)
plt.ylim(-10, 100)
```
